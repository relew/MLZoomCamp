```python
df2 = df[df["Make"] == "Lotus"][["Engine HP","Engine Cylinders"]].drop_duplicates()
```

```python
x = df2.to_numpy()
```

```python
XTX = x.T.dot(x)
```

```python
x_invers = np.linalg.inv(XTX)
```

```python
x_invers
```

```python
y = np.array([1100,800,750,850,1300,1000,1000,1300,800])
```

```python
w = x_invers.dot(x.T) * y
w
```
